```r
##https://www.quantstart.com/articles/Serial-Correlation-in-Time-Series-Analysis/
set.seed(1)
x <- seq(1,100) + 20.0*rnorm(1:100)
set.seed(2)
y <- seq(1,100) + 20.0*rnorm(1:100)
plot(x,y)
```

```r
cov(x,y)
```

```r
cor(x,y)
```
