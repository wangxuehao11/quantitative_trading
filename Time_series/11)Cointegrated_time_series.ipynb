```r
## https://www.quantstart.com/articles/Cointegrated-Time-Series-Analysis-for-Mean-Reversion-Trading-with-R/
set.seed(123)
z <- rep(0, 1000)
for (i in 2:1000) z[i] <- z[i-1] + rnorm(1)
plot(z, type="l")
```

```r
layout(1:2)
acf(z)
acf(diff(z))
```

```r
x <- y <- rep(0, 1000)
x <- 0.3*z + rnorm(1000)
y <- 0.6*z + rnorm(1000)
layout(1:2)
plot(x, type="l")
plot(y, type="l")
```

```r
comb <- 2*x - y
layout(1:2)
plot(comb, type="l")
acf(comb)
```

```r
require(tseries) 
adf.test(comb)
```

```r
pp.test(comb)
```

```r
po.test(cbind(2*x,-1.0*y))
```

```r
badcomb <- -1.0*x + 2.0*y
layout(1:2)
plot(badcomb, type="l")
acf(diff(badcomb))
adf.test(badcomb)
```
